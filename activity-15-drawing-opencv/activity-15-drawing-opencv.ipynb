## 📦 Setup: Verify Sample Image

The sample image is already included in this template folder. Let's verify it exists:

```python
# =============================================================================
# 📦 Setup: Verify Sample Image - Works Locally, Browser Colab, AND VS Code + Colab!
# =============================================================================
import os
import urllib.request

# GitHub URL for sample file
FILES = {
    'stevejob.png': 'https://raw.githubusercontent.com/KCW89/telebort-public-datasets/main/AI-2/images/stevejob.png'
}

print("Checking sample image...")
for filename, url in FILES.items():
    if os.path.exists(filename):
        size = os.path.getsize(filename) / 1024  # KB
        print(f"  ✅ {filename} ({size:.1f} KB)")
    else:
        print(f"  📥 Downloading {filename}...")
        urllib.request.urlretrieve(url, filename)
        size = os.path.getsize(filename) / 1024
        print(f"  ✅ {filename} downloaded ({size:.1f} KB)")

print("\n✅ Sample image ready!")
```

## 📚 Section 1: Loading Images - Working Example

```python
# WORKING EXAMPLE 1: Load and display an image
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image
image = cv.imread("stevejob.png")

# Display image information
print("Image Properties:")
print(f"  Shape: {image.shape}")
print(f"  Height: {image.shape[0]} pixels")
print(f"  Width: {image.shape[1]} pixels")
print(f"  Channels: {image.shape[2]} (BGR)")

# Display the original image
print("\nOriginal Image:")
cv2_imshow(image)
print("\n✅ Image loaded successfully!")
```

## 📚 Section 2: Drawing Rectangles - Working Example

```python
# WORKING EXAMPLE 2: Draw rectangles on an image
import cv2 as cv
from google.colab.patches import cv2_imshow

# Load fresh copy of image
image = cv.imread("stevejob.png")

# Draw a blue rectangle from (0,0) to (250,250)
# Parameters: image, start_point, end_point, color (BGR), thickness
cv.rectangle(image, (400, 0), (800, 500), (255, 0, 0), thickness=2)

# Draw a green filled rectangle
cv.rectangle(image, (300, 50), (450, 150), (0, 255, 0), thickness=-1)  # -1 = filled

print("Rectangles drawn:")
print("  - Blue outline rectangle (top-left corner)")
print("  - Green filled rectangle")
cv2_imshow(image)
print("\n✅ Rectangles drawn successfully!")
```

---

## 📚 Section 3: Drawing Circles and Lines - Working Example

```python
# WORKING EXAMPLE 3: Draw circles and lines
import cv2 as cv
from google.colab.patches import cv2_imshow

# Load fresh copy of image
image = cv.imread("stevejob.png")

# Calculate image center
center_x = image.shape[1] // 2
center_y = image.shape[0] // 2

# Draw a green circle at the center
# Parameters: image, center, radius, color (BGR), thickness
cv.circle(image, (center_x, center_y), 100, (0, 255, 0), thickness=4)

# Draw a red line from top-left to center
# Parameters: image, start_point, end_point, color (BGR), thickness
cv.line(image, (100, 0), (center_x, center_y), (0, 0, 255), thickness=4)

# Draw a yellow diagonal line
cv.line(image, (0, 0), (image.shape[1], image.shape[0]), (0, 255, 255), thickness=2)

print("Shapes drawn:")
print(f"  - Green circle at center ({center_x}, {center_y})")
print("  - Red line from (100, 0) to center")
print("  - Yellow diagonal line")
cv2_imshow(image)
print("\n✅ Circles and lines drawn successfully!")
```

## 📚 Section 4: Adding Text - Working Example

```python
# WORKING EXAMPLE 4: Add text annotations
import cv2 as cv
from google.colab.patches import cv2_imshow

# Load fresh copy of image
image = cv.imread("stevejob.png")

# Add text to the image
# Parameters: image, text, position, font, scale, color (BGR), thickness
cv.putText(
    image, 
    "Steve Jobs", 
    (50, 50),  # Position (x, y)
    cv.FONT_HERSHEY_COMPLEX,  # Font style
    1.5,  # Font scale
    (255, 255, 0),  # Cyan color (BGR)
    3  # Thickness
)

# Add another text with different style
cv.putText(
    image,
    "Innovation",
    (50, image.shape[0] - 50),  # Bottom-left
    cv.FONT_HERSHEY_SIMPLEX,
    2.0,
    (0, 255, 255),  # Yellow (BGR)
    4
)

print("Text added:")
print("  - 'Steve Jobs' at top-left (cyan)")
print("  - 'Innovation' at bottom-left (yellow)")
cv2_imshow(image)
print("\n✅ Text annotations added successfully!")
```
